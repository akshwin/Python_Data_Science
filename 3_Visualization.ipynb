CREATE MULTPLE PLOT WITH SUBPLOTS

```python
names=['John','Sanjay','Ellen','Sandra','Nancy','Shubham','Philip']
scores = [10,15,20,30,40,50,100]

plt.figure(figsize=(10,5))

plt.subplot(231) # row column first position
plt.bar(names,scores)

plt.subplot(232)
plt.scatter(names,scores)

plt.subplot(233)
plt.plot(names,scores)

plt.subplot(234)
plt.bar(names,scores)

plt.suptitle('Catergorail Plotting')
plt.show()
```

SEABORN

```python
import seaborn as sns
```

```python
ourdata = pd.read_csv("pokemon.csv")
```

```python
ourdata.head()
```

```python
sns.lmplot(x='sp_attack',y='sp_defense',data=ourdata)
plt.show()
```

```python
sns.lmplot(x='sp_attack',y='sp_defense',data = ourdata,fit_reg=False) # no regressiion line
```

```python
sns.lmplot(x='sp_attack',y='sp_defense',data = ourdata,fit_reg=False,hue='is_legendary')
```

```python
ourdata['is_legendary'].value_counts()
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])

x = np.arange(1,10)
a1.plot(x,np.exp(x),'b') # b is for blue

a1.set_title('range of numbers')

plt.ylim(0,1000)
plt.xlim(1,10)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=ourdata)
```

```python
ourdata1=ourdata.drop(['is_legendary','generation'],axis=1)
ourdata1.head()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=ourdata1)
```

```python
corr = ourdata1.corr() # to find correlation
corr
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(corr)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(corr,annot=True)
```

UNIVARIATE

DISTPLOT - DISTRIBUTION OF THE DATA

```python
sns.distplot(ourdata1['sp_defense'])
```

```python
sns.boxplot(ourdata1['sp_defense']) # check the above graph
```

```python
sns.distplot(ourdata1['sp_defense'],kde=False) # no curve
```

```python
sns.kdeplot(ourdata1['sp_defense'])
plt.show()
```

```python
sns.kdeplot(ourdata1['sp_defense'],shade=True)
plt.show()
```

BIVARIATE DISTRIBUTIIONS

```python
ourdata.head()
```

```python
sns.jointplot(ourdata['sp_defense'])
```

```python
sns.jointplot(data=ourdata,x='sp_defense',y='sp_attack')
```

```python
# check the correlation

ourdata1[['sp_defense','sp_attack']].corr()
```

```python
sns.jointplot(data=ourdata,x='sp_defense',y='sp_attack',kind='kde',fill=True)
plt.show()
```

```python
sns.jointplot(data=ourdata,x='sp_defense',y='sp_attack',kind='reg')
plt.show()
```

VISUALISING MORE THAN TWO VARIABLES PAIRWISE BIVARIATE DISTRIBUTIONS USING PAIRPLOT()

```python
sns.pairplot(ourdata[['sp_defense','sp_attack','speed']],kind='scatter')
plt.show()
```

CHANGE THE DIAGONALS

```python
sns.pairplot(ourdata[['sp_defense','sp_attack','speed']],kind='scatter',diag_kind='kde')
plt.show()
```
